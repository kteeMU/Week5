```python
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
    
%matplotlib inline
```

```python
shop_data = Path("shop.csv")
shop_data = pd.read_csv(shop_data, index_col="Date",infer_datetime_format=True, parse_dates=True)
shop_df = shop_data
shop_data.head()
```

```python
shop_df = shop_df.sort_index()
shop_df.head()
```

```python
#plot 
shop_df.plot()
```

```python
#calculate 20 day simple moving average
shop_sma_20 = shop_df.rolling(window=20).mean()
shop_sma_20
```

```python
shop_sma_20.plot()
```

```python
#calculate 50 day simple moving average
shop_sma_50 = shop_df.rolling(window=50).mean()
shop_sma_50.plot()
```

```python
#calculate 100 day simple moving average
shop_sma_100 = shop_df.rolling(window=100).mean()
shop_sma_100.plot()
```

```python
#overlay daily closing
ax = shop_df.plot(figsize = [25,10])

#overlay SMA20 50 and 100
shop_sma_20.plot(ax=ax)
shop_sma_50.plot(ax=ax)
shop_sma_100.plot(ax=ax)

#set the legend of the figure 
ax.legend(["Original", "SMA20", "SMA50", "SMA100"]);
```

```python
#calculate 20 day rolling standard deviation
shop_std_20 = shop_df.rolling(window=20).std()
shop_std_20.plot()
```

```python
#calculate 50 day rolling standard deviation
shop_std_50 = shop_df.rolling(window=20).std()
shop_std_50.plot()
```

```python
#calculate 100 day rolling standard deviation
shop_std_100 = shop_df.rolling(window=20).std()
shop_std_100.plot()
```

```python
#overlay daily closing
ax = shop_df.plot(figsize = [25,10])

#overlay STD20 50 and 100
shop_df.rolling(window=20).std().plot(ax=ax)
shop_df.rolling(window=50).std().plot(ax=ax)
shop_df.rolling(window=1000).std().plot(ax=ax)


#set the legend of the figure 
ax.legend(["Original", "STD20", "STD50", "STD100"]);
```
